```python
#Create function for plotting geom bar plot

def gg_barplot(data = churn_data,x = ""):
    plotresult = (
        ggplot(data) +
        aes(x = x, fill = "Churn") +
        geom_bar(position = "dodge") +
        coord_flip()
    )

    return plotresult
```

```python
gg_barplot(churn_data,x = "PhoneService")
```

There are more customers that has a phone service. Pattern for both class in terms of Churn count is the same

```python
gg_barplot(x = "MultipleLines")
```

We have more customers that did not churn than customers that did churn on all class of Multiple Lines

```python
gg_barplot(x = "InternetService")
```

Even though the pattern seems to be the same on all classes, we can see that on Fiber Optic, there is a lot churners recorded.

```python
gg_barplot(x = "OnlineSecurity")
```

We have more churns on Customers that do not have OnlineSecurity on their subscription
